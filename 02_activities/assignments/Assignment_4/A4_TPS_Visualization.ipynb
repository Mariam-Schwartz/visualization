```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = TPS_data + "/TPS Budget by command.csv"
data = pd.read_csv(file_path)

# Preprocess data for the bar chart
bar_data = data.groupby(['Year', 'Type_of_Metric'])['Amount'].sum().unstack()

# Create the bar chart
plt.figure(figsize=(10, 6))
bar_data.plot(kind='bar', figsize=(10, 6), width=0.8)
plt.title('Comparison of Actual Expenditures vs Approved Budget by Year')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.xticks(rotation=0)
plt.legend(title='Type of Metric', loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Preprocess data for the stacked bar chart
stacked_data = data[data['Type_of_Metric'] == 'Actual Expenditures']

# Define the desired order of categories
category_order = ['Salaries', 'Benefits', 'Premium Pay', 'Materials & Supplies', 'Equipment']

# Group and pivot data
stacked_grouped = stacked_data.groupby(['Year', 'Category'])['Amount'].sum().unstack()

# Reorder the categories in the DataFrame
stacked_grouped = stacked_grouped[category_order]  # Reindex columns based on category_order

# Create the stacked bar chart
plt.figure(figsize=(12, 8))
stacked_grouped.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='tab20')

plt.title('Breakdown of Actual Expenditures by Category Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Amount ($)', fontsize=12)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
