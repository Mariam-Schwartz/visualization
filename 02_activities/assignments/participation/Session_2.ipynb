```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import scipy 

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
#Make a line plot with both of variables on the same axes
fig, ax = plt.subplots(figsize=(5, 3))
#Add a legend
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
    frameon = True, #add frame around the legend
    fontsize = 12, #change font size
    ncol = 2, #specify number of columns
    shadow = True )
fig.show()
```

```python
#Make a line plot with both of variables on the same axes
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
#Moving the legend outside of the plot area
ax.legend(loc="upper left",
    bbox_to_anchor =(1, 0.5)) #specifying where we want to anchor
fig.show()
```

```python
#Annotations with the text() function
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important!", #specify coordinates and text    
    ha='center', ##alignment of text
    color = 'red', ##modify font colour
    size = 20) ##modify font size

fig.show()
```

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
#Position annotation with reference to the data, the axes, or the figure 
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData )
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes )
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure )
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
#specify the location to point at and the location where we want text
ax.annotate('This is important!', xy=(10, 95), 
    xytext=(20, 94),
    arrowprops=dict(facecolor='hotpink', arrowstyle = "wedge"))
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
#Removing tick marks and labels
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
#define a maximum number of ticks
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
#set tick marks at a particular interval 
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(12))
```

```python
#rotate axis labels and horizontal alignment, use xticks()
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")
plt.xticks(rotation=45, ha='right', fontsize = 7, font = 'serif')
```

```python
#Subplots
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])

plt.style.use('classic')
```

```python
#Create  two plots next to each other
fig,(ax1, ax2) = plt.subplots(ncols=2,
    nrows=1,
    figsize=(10, 5))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
fig.show()
```

```python
#Customizing the plots
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7, 3))

# Scatter Plot Customization
ax1.scatter(
    x1, y1, 
    c=y1, cmap='plasma', alpha=0.8, edgecolor='black', s=60
)
ax1.set_title("Scatter Plot", fontsize=10, fontweight='bold', color="#4A4A4A")
ax1.set_xlabel("Xxxx", fontsize=8)
ax1.set_ylabel("Yyyy", fontsize=8)
ax1.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

# Bar Plot Customization
bars = ax2.bar(
    x2, y2, 
    color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c44d56'], 
    #edgecolor="black", linewidth=1.2
)
ax2.set_title("Bar Plot", fontsize=10, fontweight='bold', color="#4A4A4A")
ax2.set_xlabel("Categories", fontsize=8)
ax2.set_ylabel("Values", fontsize=8)
ax2.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, alpha=0.7)

plt.show()
```

```python
plt.style.use('Solarize_Light2')
#Subplots without a grid 
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7, 4),
                                    layout="constrained")
fig.patch.set_facecolor('white')

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

#Add x axis titles
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```
