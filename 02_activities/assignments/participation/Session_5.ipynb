```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
# Create a line plot using Seaborn
tipgraph = sns.lineplot(
    data=tips,          # Dataset to use
    x='total_bill',     # Column to use for the x-axis
    y='tip'             # Column to use for the y-axis
)

# Customize the graph with a title and axis labels
tipgraph.set(
    title='Tip vs. Total Bill',  # Set the title of the plot
    xlabel='Total Bill ($)',     # Set the label for the x-axis
    ylabel='Tip Amount ($)'      # Set the label for the y-axis
)
```

```python
# Create a figure with specified size
fig = plt.subplots(figsize=(10, 3))

# Create a customized line plot using Seaborn
tipgraph = sns.lineplot(
    data=tips,                 # Dataset to use 
    x='total_bill',            # Column for the x-axis
    y='tip',                   # Column for the y-axis
    color='hotpink',           # Set the line color to hot pink
    linestyle='--',            # Set the line style to dashed
    linewidth=3,               # Set the line thickness to 3
    marker='o',                # Add circular markers to data points
    markerfacecolor='indigo'   # Set the marker fill color to indigo
)

# Customize the graph with a title and axis labels
tipgraph.set(
    title='Tip vs. Total Bill',  # Set the title of the plot
    xlabel='Total Bill ($)',     # Set the label for the x-axis
    ylabel='Tip Amount ($)'      # Set the label for the y-axis
)
```

```python
tipgraph = sns.scatterplot(data = tips,
                           x = 'total_bill',
                           y = 'tip',
                           style = 'time', # Different marker styles based on the 'time' column
                           hue = 'day',      # Different colors based on the 'day' column
                           palette = ['purple','hotpink','deepskyblue','yellowgreen']) #custom color palette

tipgraph.set(title = 'Tip vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip AMount ($)')
```

```python
# Create a pair plot using Seaborn
sns.pairplot(data = tips,
             hue ='day')
```

```python
# Create a relational plot using Seaborn
daysplot = sns.relplot(data=tips,
                       x="total_bill",
                       y="tip",
                       hue="sex",
                       col="day",
                       kind="scatter",
                       col_wrap=2)

daysplot
```

```python
import plotly.graph_objects as go

x1 = np.array(["Luffy","Zoro","Nami","Usopp","Snaji"])
y1 = np.array([110,180,240,99,220])
```

```python
#generate an interactive bar chart using Plotly
graph = go.Figure()

graph.add_trace(go.Bar(x=x1,y=y1))

graph.update_layout(title = "Pirate Scores",
                    xaxis_title = "Pirates",
                    yaxis_title = "Score")
```

```python
graph = go.Figure()

graph.add_trace(go.Scatter(
    x= x1,
    y= y1,
    mode = 'markers',
    marker = dict(
        size =15,
        color = 'hotpink',
        opacity = 1,
        line = dict(width = 5,
                    color = 'purple')
    )
))

graph.update_layout(title='Interactive Pirate Plot',
                    xaxis_title = 'Pirates',
                    yaxis_title = 'Scores',
                    width = 500,
                    height = 500)
```

```python
from matplotlib_venn import venn2,venn2_circles, venn2_unweighted

# Define set A with a list of items
A = set(['apple','banana','watermelon'])
B= set(['pumpkin','blueberry','apple','key lime'])

# Create a Venn diagram with the two sets
diagram = venn2([A,B],
                set_labels = ('Fruits', 'Pies'),
                set_colors = ('blue', 'red'),
                alpha = 0.5)
# Customize the label for the region representing A - B, A & B, and B - A
diagram.get_label_by_id("10").set_text("\n".join(A - B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B - A))
```
